For a higher class corresponds a higher fare(obvious), so they were more likely to survive, 30 - 870 for the first class in comparison with 3 - 7 for the last one.

```python
df_female = df[df['Sex'] == 'female']
```

```python
pd.DataFrame(np.abs(df_female.corr(numeric_only=True)['Survived'])).sort_values(by='Survived', ascending=False)
```

Interestingly, for females the number of siblings/spouses on the board was important.

```python
pd.concat([df_female[['Survived','SibSp']].groupby(by=['SibSp']).mean(), df_female[['Survived','SibSp']].groupby(by=['SibSp']).count()], axis=1)
```

As we see, the bigger amount of siblings/spouces a woman has, the lower chances to survive. My assumption is that the number 1-2 meant that they weren't caring for a large group of people. For those with a larger number, I suppose they were kids, and their parents hadn't managed to help all of them. The majority of the families with a large number of kids were emigrants, so the class filter still applies.

```python
pd.DataFrame(np.abs(df_female.corr(numeric_only=True)['Age'])).sort_values(by='Age', ascending=False)
```

```python
df_female[['Survived','Age']].groupby(pd.cut(df_female['Age'], bins=10)).mean().sort_values(by='Survived', ascending=False)
```

```python
df_male = df[df['Sex'] == 'male']
```

```python
pd.DataFrame(np.abs(df_male.corr(numeric_only=True)['Survived'])).sort_values(by='Survived', ascending=False)
```

Surprisingly, for males wasn't really important. But still, the class and fare were of top importance.

This drives us to the thoughts that actually the sex might be a decisive factor for survival.

Let's encode the Sex: male = 1, female = 0 (as the data is clean and doesn't have NaNs)

```python
np.unique_values(df['Sex'])
```

```python
df['Sex'] = df['Sex'] == 'male'
```

```python
pd.DataFrame(np.abs(df.corr(numeric_only=True)['Survived'])).sort_values(by='Survived', ascending=False)
```

Actually, we figured out that sex was the most important factor for survival.
